```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('labeled_data.csv')
prof = pd.read_csv('profanity_en.csv')
```

```python
df.head()
```

```python
prof.head()
```

```python
df.drop(['Unnamed: 0', 'count',	'hate_speech',	'offensive_language',	'neither'], axis=1, inplace=True)
df.head()
```

```python
prof.drop(['canonical_form_1',	'canonical_form_2',	'canonical_form_3',	'category_1',	'category_2',	'category_3',	'severity_rating',	'severity_description'], axis=1, inplace=True)
prof.head()
```

```python
prof['class'] = 0
```

```python
prof.head()
```

```python
df['text'] = df['tweet']
df.drop('tweet', axis=1, inplace=True)
df.head()
```

```python
df = pd.concat([df, prof], ignore_index=True)
df
```

```python
df['profanity'] = [p['text'] if p['text'] in text else None for text in df['text'] for _, p in prof.iterrows()]
```

```python
!pip install flashtext
from flashtext import KeywordProcessor

keyword_processor = KeywordProcessor()
keyword_processor.add_keywords_from_list(prof['text'].tolist())
df['profanity'] = [keyword_processor.extract_keywords(text) for text in df['text']]
```
